```python
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")
```

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}
```

```python
prompt = "The future of Artificial Intelligence is"

for name, model in models.items():
    print(f"\n=== {name} ===")
    try:
        generator = pipeline("text-generation", model=model)
        output = generator(prompt, max_length=30)
        print(output[0]["generated_text"])
    except Exception as e:
        print("FAILED:", e)
```

```python
masked_sentence = "The goal of Generative AI is to [MASK] new content."

for name, model in models.items():
    print(f"\n=== {name} ===")
    try:
        fill = pipeline("fill-mask", model=model)
        outputs = fill(masked_sentence)
        for o in outputs[:3]:
            print(o["token_str"], " | score:", round(o["score"], 4))
    except Exception as e:
        print("FAILED:", e)
```

```python
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."
question = "What are the risks?"

for name, model in models.items():
    print(f"\n=== {name} ===")
    try:
        qa = pipeline("question-answering", model=model)
        result = qa(question=question, context=context)
        print("Answer:", result["answer"])
        print("Score:", round(result["score"], 4))
    except Exception as e:
        print("FAILED:", e)
```

## Observation Table – Model Benchmark Results

| Task | Model | Result | Observation | Architectural Reason |
|------|-------|--------|-------------|----------------------|
| **Text Generation** | **BERT** | Failure | Generated incoherent or repetitive text | Encoder-only model; not trained for autoregressive generation |
| | **RoBERTa** | Failure | Could not generate meaningful continuation | Encoder-only architecture with no decoder |
| | **BART** | Success | Generated fluent and contextually relevant text | Encoder–Decoder architecture designed for generation |
| **Fill-Mask** | **BERT** | Success | Predicted correct words like *create*, *generate* | Trained using Masked Language Modeling (MLM) |
| | **RoBERTa** | Success | Accurate predictions using `<mask>` token | Optimized MLM pretraining |
| | **BART** | Partial | Predictions reasonable but less confident | Not primarily trained for MLM tasks |
| **Question Answering** | **BERT** | Partial | Relevant but incomplete answers | Base model not fine-tuned for QA |
| | **RoBERTa** | Partial | Slightly better consistency than BERT | Stronger pretraining, still not QA-tuned |
| | **BART** | Partial | Verbose or imprecise answers | Generative bias without QA fine-tuning |
